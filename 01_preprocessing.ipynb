```python
# KETI txt를 로드하는 함수
import re

def load_keti_data(path):
    with open(path, "r", encoding="utf-8") as f:    # 한글 깨짐 방지
        return f.read()
```

```python
# 실제 파일 읽기
path = "KETI_대화데이터_일상_오피스.txt"  # 함수 호출 -> 실제 데이터 읽기
raw_text = load_keti_data(path)

print(raw_text[:200])   # 날 것 텍스트 
```

```python
# 문장만 추출
import pandas as pd

lines = raw_text.splitlines()   # 한 발화로 라인 분리

corpus = pd.Series( # 말뭉치 생성
    [line.split("\t", 1)[1].strip() for line in lines if "\t" in line],
    #if "\t" in line: 탭 안에 있는 줄 처리 
    #line.split("\t", 1): 탭 기준 최대 2개 쪼개는 거
    #strip(): 앞 뒤 공백 제거
    #pd.Series(): 문장 목록 판다스 시리즈로 변환 -> map, filna()로 처리
    name="sentence"
)

corpus.head(10)
```

```python
#텍스트 정제
def clean_text(text):   # 노이즈 제거하기
    text = str(text)

    # 특수문자 제거 (한글/영문/숫자/공백만 남김)
    text = re.sub(r"[^0-9A-Za-z가-힣\s]", " ", text)    #re.sub(허용문자 제외, " ", text) :한글/영문/숫자/공백만 남기고 나머지 제거

    return text

corpus_clean = corpus.fillna("").astype(str).map(clean_text)
                        #fillna("") : NaN이면 빈 문자열로
                        #astype(str) : 전부 문자열로 통일
                        #map(clean_text) : 문장 하나하나에 clean_text 적용
corpus_clean.head(10)
```

```python
from kiwipiepy import Kiwi  #Kiwi 형태소 분석기 객체 생성
kiwi = Kiwi()

def tokenize_meaningful(text):  #kiwi.tokenize(text) : 형태소 분석 결과 리스트 반환
    tokens = []
    for token in kiwi.tokenize(text):
        # 명사(N), 동사(V), 외래어(SL) 중심
        if token.tag.startswith(("N", "V", "SL")) and len(token.form) > 1:  #N: 명사류/V: 동사/용언 계열/SL: 외래어
            tokens.append(token.form)                                       #너무 짧은 토큰(한 글자) 제거해서 노이즈 감소
    return tokens

tokenized = corpus_clean.map(tokenize_meaningful)   #문장별 토큰 리스트 만들기

tokenized.head()
```

```python
#vocabulary 만들기
from collections import Counter #단어 빈도 세기(= vocabulary)

#all_tokens=문장별 토큰 리스트들을 1차원으로 펼침(flatten)
all_tokens = [token for sent in tokenized for token in sent]

word_counts = Counter(all_tokens)       #Counter(all_tokens)=단어별 등장 횟수 딕셔너리 생성

# 2회 이상 등장하는 단어만
#word_counts.items():(단어, 빈도) 묶음
#count >= 2 조건으로 필터링
vocab = [word for word, count in word_counts.items() if count >= 2] #단어장 만들기

len(vocab)
```

```python
word_counts.most_common(20) #most_common() — 결과 검증(상위 단어 확인)
# 상위 단어 20개 확인
```
